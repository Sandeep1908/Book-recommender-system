# Popularity based

```python
rating_with_name=new_rating_df.groupby('Book-Title').count()['Book-Rating'].reset_index()
rating_with_name.rename(columns={'Book-Rating':'no_of_rating'},inplace=True)
rating_with_name
```

```python
avg_rating=new_rating_df.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating
```

```python
popular_df=rating_with_name.merge(avg_rating,on='Book-Title')
```

```python
popular_df=popular_df[popular_df['no_of_rating']>250]
```

```python
top_50_books=popular_df.sort_values('avg_rating',ascending=False).head(50)
```

```python
popularity=book.merge(top_50_books,on='Book-Title')
```

```python
popularity
```

# Collabrative based system

```python
new_rating_df.head(2)
```

```python
x=new_rating_df.groupby('User-ID').count()['Book-Rating'] >200
```

```python
users=x[x].index
```

```python
filter_rating=new_rating_df[new_rating_df['User-ID'].isin(users)]
```

```python
filter_rating.shape
```

```python
y=filter_rating.groupby('Book-Title').count()['Book-Rating'] >=50
```

```python
famous_books=y[y].index
```

```python
final_rating=filter_rating[filter_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating.head()
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(pt)
```

```python
def recommended(book_name):
    try:
        idx=np.where(pt.index==book_name)[0][0]
        book_idx=sorted(list(enumerate(similarity[idx])),key=lambda x:x[1],reverse=True)[1:10]

        for i in book_idx:
            print(pt.index[i[0]])
    except:
        print('Book Not Found')
```

```python
recommended('Harry poter')
```
